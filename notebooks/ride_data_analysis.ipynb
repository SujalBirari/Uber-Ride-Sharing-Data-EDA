```python
cat_null_cols = ["Reason for cancelling by Customer", "Driver Cancellation Reason", "Incomplete Rides Reason", "Payment Method"]
num_null_cols = ["Cancelled Rides by Customer", "Cancelled Rides by Driver", "Incomplete Rides", "Booking Value", "Ride Distance", "Driver Ratings", "Customer Rating"]
(df.isna().mean()*100).sort_values(ascending=False)
```

```python
df.head(10)
```

```python
for col in cat_null_cols:
    df[col] = df[col].fillna('Unknown')
df.head()
```

```python
df['Vehicle Type'].value_counts()
```

```python
for col in num_null_cols:
    df[col] = df[col].fillna(df.groupby(['Vehicle Type'])[col].transform('median'))

for col in ["Avg VTAT", "Avg CTAT"]:
    df[col] = df[col].fillna(df.groupby(['Vehicle Type'])[col].transform('mean'))
df.head(10)
```

# B. Demand and Supply Patterns

1. How do total bookings vary by month, day of week, and hour of day?

```python
df['Month']      = df['Datetime'].dt.month  
df['DayOfWeek']  = df['Datetime'].dt.day_name()
df['Hour']       = df['Datetime'].dt.hour

bookings_by_month = df.groupby('Month').size().sort_index()

bookings_by_day = (
    df.groupby('DayOfWeek').size().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
)

bookings_by_hour = df.groupby('Hour').size().sort_index()

print(bookings_by_month)
print(bookings_by_day)
print(bookings_by_hour)
```

2. Which vehicle types see the highest demand and when?

```python
df.head()
```

```python
completed_rides = df[df['Booking Status'] == 'Completed']

vehicle_counts = completed_rides['Vehicle Type'].value_counts()
print(vehicle_counts)
```

```python
hourly_demand = completed_rides.groupby(['Vehicle Type', 'Hour']).size().reset_index(name='Count')
hourly_demand
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=hourly_demand, x='Hour', y='Count', hue='Vehicle Type')
plt.title("Hourly Demand for Each Vehicle Type")
plt.xlabel("Day of Week")
plt.ylabel("Number of Completed Rides")
plt.show()
```

3. Are there specific pickup locations that act as hotspots for demand?

```python
pickup_counts = completed_rides['Pickup Location'].value_counts().reset_index()
pickup_counts.columns = ['Pickup Location', 'Ride Count']
```

```python
df.head()
```

```python
top_locations = pickup_counts.head(20)  # Top 20 hotspots

plt.figure(figsize=(12,6))
sns.barplot(data=top_locations, y='Pickup Location', x='Ride Count', hue='Pickup Location', palette='viridis')
plt.title("Top 20 Pickup Locations (Demand Hotspots)")
plt.xlabel("Number of Completed Rides")
plt.ylabel("Pickup Location")
plt.show()
```

```python
hourly_hotspots = completed_rides.groupby(['Pickup Location', 'Hour']).size().reset_index(name='Count')
top_hotspots = hourly_hotspots[hourly_hotspots['Pickup Location'].isin(top_locations['Pickup Location'])]

heatmap_data = top_hotspots.pivot(index='Pickup Location', columns='Hour', values='Count')  # âœ…

plt.figure(figsize=(12,8))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlOrRd")
plt.title("Hourly Demand at Top Pickup Locations")
plt.show()
```

# 3. Ride Success and Cancellations

1. What is the overall completion vs. cancellation rate, and how do these vary:

by vehicle type

by time of day/day of week

by pickup/drop region

```python
total_counts = df['Booking Status'].value_counts()
completion_rate = total_counts.get('Completed',0) / total_counts.sum() * 100
cancellation_rate = 100 - completion_rate

print(f"Overall Completion Rate: {completion_rate:.2f}%")
print(f"Overall Cancellation Rate: {cancellation_rate:.2f}%")
```

```python
vehicle_status = df.groupby(['Vehicle Type', 'Booking Status']).size().reset_index(name='Count')
vehicle_total = vehicle_status.groupby('Vehicle Type')['Count'].transform('sum')
vehicle_status['Rate (%)'] = vehicle_status['Count'] / vehicle_total * 100

vehicle_status
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=vehicle_status, x='Vehicle Type', y='Rate (%)', hue='Booking Status')
plt.title("Completion vs Cancellation Rate by Vehicle Type")
plt.ylabel("Rate (%)")
plt.xlabel("Vehicle Type")
plt.show()
```

# Financial and Revenue Insights

```python
df.head()
```

1. Total revenue and average fare per ride.

```python
total_revenue = df['Booking Value'].sum()
average_fare = df['Booking Value'].mean()

print(f"Total Revenue: â‚¹{total_revenue:,.2f}")
print(f"Average Fare per Ride: â‚¹{average_fare:,.2f}")
```

2. Contribution of each payment method to total revenueâ€”any growth in UPI/Cashless usage?

```python
payment_revenue = df.groupby('Payment Method')['Booking Value'].agg(['sum','mean']).reset_index()
payment_revenue.columns = ['Payment Method', 'Total Revenue', 'Average Fare']
payment_revenue
```

```python
# Total revenue per payment method
plt.figure(figsize=(12,6))
sns.barplot(data=payment_revenue, x='Payment Method', y='Total Revenue', hue='Total Revenue', palette='viridis', legend=False)
plt.title("Total Revenue by Payment Method")
plt.ylabel("Revenue (â‚¹ in Crores)")
plt.xlabel("Payment Method")
plt.show()

# Average fare per payment method
plt.figure(figsize=(12,6))
sns.barplot(data=payment_revenue, x='Payment Method', y='Average Fare', hue='Average Fare', palette='magma', legend=False)
plt.title("Average Fare per Ride by Payment Method")
plt.ylabel("Average Fare (â‚¹)")
plt.xlabel("Payment Method")
plt.show()
```

```python
# Growth Trend in Cash or UPI

df['Month'] = df['Datetime'].dt.to_period('M').astype(str)  # Monthly granularity

upi_trend = df[df['Payment Method'] == 'UPI'].groupby('Month')['Booking Value'].sum().reset_index()
cash_trend = df[df['Payment Method'].isin(['Debit Card','Credit Card','Uber Wallet'])].groupby('Month')['Booking Value'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=upi_trend, x='Month', y='Booking Value', marker='o', label='UPI')
sns.lineplot(data=cash_trend, x='Month', y='Booking Value', marker='o', label='Other Cashless')
plt.title("Monthly Revenue Trend for UPI vs Other Cashless Payments")
plt.xlabel("Month")
plt.ylabel("Revenue (â‚¹)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

3. Which vehicle types drive the highest revenue per km?

```python
completed_rides = df[df['Booking Status'] == 'Completed'].copy()

completed_rides['Revenue_per_km'] = completed_rides['Booking Value'] / completed_rides['Ride Distance']

rev_per_km_vehicle = completed_rides.groupby('Vehicle Type')['Revenue_per_km'].mean().reset_index()
rev_per_km_vehicle.columns = ['Vehicle Type', 'Avg Revenue per km']

rev_per_km_vehicle
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=rev_per_km_vehicle.sort_values('Avg Revenue per km', ascending=False), 
            x='Vehicle Type', 
            y='Avg Revenue per km',
            hue='Avg Revenue per km',
            legend=False
)
plt.title("Average Revenue per km by Vehicle Type")
plt.ylabel("Revenue per km (â‚¹)")
plt.xlabel("Vehicle Type")
plt.show()
```
