```python
# Handling date-time anamolies

# 1. Combining date and time into one
df['Datetime'] = pd.to_datetime(
    df['Date'].astype(str)+' '+df['Time'].astype(str),
    errors='coerce'
)
df.head()
```

```python
# Checking for future booking anamolies

now = pd.Timestamp.now()
future_rows = df[df['Datetime'] > now]
future_rows
```
